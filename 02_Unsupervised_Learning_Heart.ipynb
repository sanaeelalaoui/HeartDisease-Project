# Part 02: Unsupervised Learning - Heart Disease

This notebook covers dimensionality reduction and clustering techniques applied to the Heart Disease dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA, NMF
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
```

## 1. Load Preprocessed Data

```python
df = pd.read_csv('heart_preprocessed.csv')
X = df.drop('target', axis=1)
y = df['target']
```

## 2. Dimensionality Reduction (PCA, t-SNE, NMF)

```python
# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# NMF (Requires non-negative inputs, we use absolute values of scaled data or original imputed data)
nmf = NMF(n_components=2, init='random', random_state=42)
X_nmf = nmf.fit_transform(np.abs(X)) # Simplified approach for demo

# Plot PCA
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='viridis')
plt.title('PCA of Heart Disease Data')
plt.show()
```

## 3. Clustering (K-Means, Agglomerative, DBSCAN)

```python
# K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
clusters_km = kmeans.fit_predict(X)

# Agglomerative
agglo = AgglomerativeClustering(n_clusters=2)
clusters_agg = agglo.fit_predict(X)

# DBSCAN
dbscan = DBSCAN(eps=3, min_samples=5)
clusters_db = dbscan.fit_predict(X)

print("Clustering complete.")
```
